```python
success=list()
for i in range(0,len(p3p_s)):
    if(p3p_s[i]!=999):
        if((float)(p3p_s[i])>35):
            success.append(1)
        else:
            success.append(0)
    else: success.append(999)
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
X = df.iloc[:,1:].values
y = df["Success"].values

XTrain, XTest, yTrain, yTest = train_test_split(X, y, random_state=1, test_size=0.35)
```

```python
gnb = GaussianNB()


# Train classifier
gnb.fit(XTrain,yTrain)

y_pred = gnb.predict(XTest)
y_pred_train = gnb.predict(XTrain)

# Print results
print('Accuracy on Train data= ', metrics.accuracy_score(y_true = yTrain, y_pred = y_pred_train))
print('Accuracy on test data= ', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
```
